```python
#sklearn.datasets for the package that creates the data
from sklearn.datasets import make_regression
#matplotlib to print out and see what the data looks like
import matplotlib.pyplot as plt
```

```python
#This represents a very basic, straight line output of the regression output
X,y = make_regression(n_samples=100, n_features=1, noise=0)
plt.scatter(X,y);
```

```python
#This shows what happens when when you add a little bit of noise to the data
X,y = make_regression(n_samples=100, n_features=1, noise=5)
plt.scatter(X,y);
```

```python
#This has the same amount of noise but has 2 features
X,y = make_regression(n_samples=100, n_features=2, noise=.5)
x1 = [i[0] for i in X]
x2 = [i[1] for i in X]
plt.scatter(x1,y);
plt.scatter(x2,y);
```

```python
#This is measuring how gaussian the data is. 
X,y = make_regression(n_samples=100, n_features=1, noise=.5)
#plt.scatter(X,y);
plt.hist(X,bins=10);
```
